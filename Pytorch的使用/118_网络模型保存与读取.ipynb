# 1. 网络模型保存(方式一)

```python
import torchvision
import torch
vgg16 = torchvision.models.vgg16(pretrained=False)
torch.save(vgg16,'D:\\深度学习\\100_土堆数据集\\model\\vgg.pth') # 保存方式一：模型结构 + 模型参数      
print(vgg16)
```

# 2. 网络模型导入(方式一)

```python
import torch
model = torch.load('D:\\深度学习\\100_土堆数据集\\model\\vgg.pth',weights_only=False) # 保存方式一对应的加载模型    
print(model)
```

# 3. 网络模型保存(方式二)

```python
import torchvision
import torch
vgg16 = torchvision.models.vgg16(pretrained=False)
torch.save(vgg16.state_dict(),'D:\\深度学习\\100_土堆数据集\\model\\vgg_2.pth') # 保存方式二：模型参数（官方推荐）,不再保存网络模型结构  
print(vgg16)
```

# 4. 网络模型导入(方式二)

```python
import torch
import torchvision
model = torch.load('D:\\深度学习\\100_土堆数据集\\model\\vgg_2.pth') # 导入模型参数   
print(model)
```

```python
import torch
import torchvision
vgg16 = torchvision.models.vgg16(pretrained=False)
print(vgg16)
vgg16.load_state_dict(torch.load("D:\\深度学习\\100_土堆数据集\\model\\vgg_2.pth")) # 将模型参数导入到模型结构中   
print(vgg16)
```
